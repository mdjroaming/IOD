```python
def count_outliers_iqr(df):
    outlier_counts = {}
    for column in df.columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        
        # Calculate IQR
        IQR = Q3 - Q1
        
        # Define outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outlier_count = df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]
        outlier_counts[column] = outlier_count
    
    return outlier_counts

# Count outliers
outlier_counts = count_outliers_iqr(df)

# Print outlier counts
for feature, count in outlier_counts.items():
    print(f"Feature: {feature} - Number of Outliers: {count}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)
```

# Step 8: Define and Fit Models

Define the model and its hyper-parameters.

Consider the parameters and hyper-parameters of each model at each (re)run and after checking the efficiency of a model against the training and test datasets.

```python
# Initialise classifiers
np.random.seed(42)
clf1 = DecisionTreeClassifier(criterion = 'entropy', max_depth = 2)
clf2 = KNeighborsClassifier(n_neighbors = 1)
clf3 = SVC()
clf4 = RandomForestClassifier(max_depth = 2)

```

```python
# Initialize Bagging Classifiers
bagging1 = BaggingClassifier(
    estimator = clf1,
    n_estimators = 10,
    max_samples = 0.8,
    max_features = 0.8)
bagging2 = BaggingClassifier(
    estimator = clf2,
    n_estimators = 10,
    max_samples = 0.8,
    max_features = 0.8)
bagging3 = BaggingClassifier(
    estimator = clf3,
    n_estimators = 10,
    max_samples = 0.8,
    max_features = 0.8)
bagging4 = BaggingClassifier(
    estimator = clf4,
    n_estimators = 10,
    max_samples = 0.8,
    max_features = 0.8)
```

```python
# Train Bagging Classifiers
bagging1.fit(X_train, y_train)
bagging2.fit(X_train, y_train)
bagging3.fit(X_train, y_train)
bagging4.fit(X_train, y_train)
```

# Step 9: Verify and Evaluate the Training Model
- Use the **training** data to make predictions
- What metrics are appropriate for the modelling approach used
- For **Supervised** models:
    - Check the **Training Results** with the **Training Predictions** during development
- Analyse, modify the parameters and hyper-parameters and repeat (within reason) until the model does not improve

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Check accuracy of fitting against the training data
# Predict on the training data
y_train_pred1 = bagging1.predict(X_train)
y_train_pred2 = bagging2.predict(X_train)
y_train_pred3 = bagging3.predict(X_train)
y_train_pred4 = bagging4.predict(X_train)

# Calculate training accuracy
training_accuracy_bagging1 = accuracy_score(y_train, y_train_pred1)
training_accuracy_bagging2 = accuracy_score(y_train, y_train_pred2)
training_accuracy_bagging3 = accuracy_score(y_train, y_train_pred3)
training_accuracy_bagging4 = accuracy_score(y_train, y_train_pred4)

print(f"Training Accuracy Bag1: {training_accuracy_bagging1:.4f}")
print(f"Training Accuracy Bag2: {training_accuracy_bagging2:.4f}")
print(f"Training Accuracy Bag3: {training_accuracy_bagging3:.4f}")
print(f"Training Accuracy Bag4: {training_accuracy_bagging4:.4f}")
```

```python
# Evaluate the classifier
print("Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred1))
print("\nClassification Report:")
print(classification_report(y_train, y_train_pred1))
```

# Step 10: Make Predictions and Evaluate the Test Model
**NOTE**: **Do this only after not making any more improvements in the model**.

- Use the **test** data to make predictions
- For **Supervised** models:
    - Check the **Test Results** with the **Test Predictions**

```python
# Make predictions
y_pred1 = bagging1.predict(X_test)
y_pred2 = bagging2.predict(X_test)
y_pred3 = bagging3.predict(X_test)
y_pred4 = bagging4.predict(X_test)
```

```python
# Evaluate each model
def evaluate_model(y_test, y_pred, model_name):
    print(f"{model_name} Performance:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

# Print evaluation results
evaluate_model(y_test, y_pred1, "Bagging with Decision Tree")
evaluate_model(y_test, y_pred2, "Bagging with KNN")
evaluate_model(y_test, y_pred3, "Bagging with SVC")
evaluate_model(y_test, y_pred4, "Bagging with Random Forest")
```

```python
# Look at the ROC curve

from sklearn import metrics  # Import the metrics module
from sklearn.metrics import roc_curve, auc

# Compute ROC curve
fpr, tpr, _= metrics.roc_curve(y_test, y_pred1)
roc_auc = metrics.auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Decision Tree')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()


```
