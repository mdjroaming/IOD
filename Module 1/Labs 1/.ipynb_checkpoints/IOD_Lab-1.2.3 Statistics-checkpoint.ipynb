````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    
    # Calculate mean of X and Y
    mean_x = sum(X) / len(X)
    mean_y = sum(Y) / len(Y)
    
    # Calculate the number of data points
    n=len(X)
    
    # Calculate the numerator and enominator in the equation
    numer = sum([xi * yi for xi,yi in zip(X,Y)]) - n * mean_x * mean_y
    denom = sum([xi **2 for xi in X]) - n * mean_x **2
    
    # Calucluate the slope (m)
    m = numer / denom
    
    # Calculate the intercept (b)
    b = mean_x - m * mean_x
    
    # Print the equation of the best-fit line
    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(b, m))
    
    return m,b
    '''
    Return m, b
    # ANSWER
    '''

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X,Y)
yfit = [b + m * xi for xi in X]
plt.plot(X, yfit);
plt.show()
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5

# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution

# ANSWER
prob = stats.probplot(x, dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
ax2 = fig.add_subplot(212)
xt, _ = stats.boxcox(x)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
````

# 13. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]

print(mean_confidence_interval(data))

results = stats.ttest_1samp(data,170,alternative='greater')
````
