# Predicting Diamond Prices

This dataset contains 10 features in which 'Price(in US dollars)' is dependent feature.

1. Carat(Weight of Diamond) -> Weight of Diamond.22. Cut(Quality) -> Quality of cut (Fair, Good, Very Good, Premium, Ideal).
 3. Color -> Diamond Color(from J -> 'worst' to D -> 'Best')
 4. Clarity -> Measurement of Transparency(how clear the Diamond is) 
                  Sequence of clarity ( I1 (worst quality), SI2, SI1, VS2, VS1, VVS2, VVS1, IF(best quality)).
 5. Table -> Width of top of a Diamnd.
 6. Price(in US dollars) -> Price of Diamond in US dolars.
 7. X(length) -> Length of Diamond i mm. 
 8. Y(width) -> Width of Diamond n mm. 
 9. Z(depth) -> Depth of Diamond in mm.
10. Depth -> Total depth percentage. It can cas follows:e formula.
                  Total Depth % = z / mean(x , y) or z * 2 / (x + y).

```python
# Create a model using the data 'as is', taking all variables as features (except Price of course)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error
```

```python
# Create a model
model = LinearRegression()
```

```python
# Fit predictor and target varables using linear regression

# define the target variable (dependent variable) as y
y = dia['price']

# Take all columns except target as predictor columns
predictor_columns = [c for c in dia.columns if c != 'price']

# Load the dataset as a pandas data frame
X = pd.DataFrame(dia, columns = predictor_columns)
```

```python
# Checking price has been excluded
X.head()
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Fit the model with the Training data
model.fit(X_train,y_train)
```

```python
# Find Coefficient and Intercept

print('Coefficient:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
# Carat (weight of diamond) has a large weighting on the price
```

```python
# Calculate the score (R^2 for Regression) for Training Data

r2_train = model.score(X_train, y_train)
print("R2 for training data:", r2_train)
```

```python
# Calculate the score (R^2 for Regression) for Testing Data

r2_test = model.score(X_test, y_test)
print("R2 for testing data:", r2_test)
```

```python
# Predict prices from X_test using original model
y_pred = model.predict(X_test)
print(y_pred [:5])
```

```python
# Evaluate the model
print('R2:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', root_mean_squared_error(y_test, y_pred))
```

```python
# Visualise the result
sns.histplot(dia.price, kde=True, label='Actual Values')
sns.histplot(y_pred, kde=True, label='Predicted Values')
plt.legend()
plt.show()
```

```python
# Hmm, model obviously not ideal as it is predicting negative values and predicted prices are much lower than the actual prices.
```

```python
# Represent this visually

def view_coeff(X, model):
    model_coefs = pd.DataFrame({'variable': X.columns,
                                'coef': model.coef_,
                                'abs_coef': np.abs(model.coef_)})
    model_coefs.sort_values('abs_coef', inplace=True, ascending=False)
    sns.barplot(x="variable", y="coef", data=model_coefs)
```

```python
# Plot Coefficients
view_coeff(X, model)
```

```python
# Define a range of alpha values (or specific values) to test
ridge_alphas = np.logspace(-6, 6, 1000)  # from 1e-6 to 1e6, 1000 values in total

# Create RidgeCV object with specified alphas and number of folds (cv)
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)

# Fit RidgeCV on training data to find the optimal alpha
optimal_ridge.fit(X_train, y_train)

# Print the optimal alpha found
print('Optimal Alpha:', optimal_ridge.alpha_)
```

```python
# Initialize Ridge regression model with a specified alpha (regularization strength)
ridge_model = Ridge(alpha=89.5)  

# Fit Ridge regression model
ridge_model.fit(X_train, y_train)

# Predict on training and test sets
y_train_ridge_pred = ridge_model.predict(X_train)
y_test_ridge_pred = ridge_model.predict(X_test)

# Evaluate model for training and test sets
train_r2_ridge = r2_score(y_train, y_train_ridge_pred)
test_r2_ridge = r2_score(y_test, y_test_ridge_pred)

train_rmse = root_mean_squared_error(y_train, y_train_ridge_pred)
test_rmse = root_mean_squared_error(y_test, y_test_ridge_pred)

print('Train R2:', train_r2_ridge)
print('Test R2:', test_r2_ridge)

print('Train RMSE:', train_rmse)
print('Test RMSE:', test_rmse)

# Print the coefficients and intercept of the Ridge regression model
print("Ridge Coefficients:", ridge_model.coef_)
print("Ridge Intercept:", ridge_model.intercept_)
```

```python
# Plot Coefficients
view_coeff(X, ridge_model)
```

```python
# Predict prices from X_test using original model
y_pred_ridge = ridge_model.predict(X_test)
print(y_pred_ridge [:5])
```

```python
# Visualise the result
sns.histplot(dia.price, kde=True, label='Actual Values')
sns.histplot(y_pred_ridge, kde=True, label='Predicted Values')
plt.legend()
plt.show()
```

```python
# Evaluate the model
print('R2_ridge:', r2_score(y_test, y_pred_ridge))
print('MAE:', mean_absolute_error(y_test, y_pred_ridge))
print('MSE:', mean_squared_error(y_test, y_pred_ridge))
print('RMSE:', root_mean_squared_error(y_test, y_pred_ridge))
```

```python
# Represent this visually

def view_coeff(X_2, linreg):
    linreg_coefs = pd.DataFrame({'variable': X_2.columns,
                                'coef': linreg.coef_,
                                'abs_coef': np.abs(linreg.coef_)})
    linreg_coefs.sort_values('abs_coef', inplace=True, ascending=False)
    sns.barplot(x="variable", y="coef", data=linreg_coefs)
```

```python
# Plot Coefficients
view_coeff(X_2, linreg)
```

```python
# Define a range of alpha values (or specific values) to test
ridge2_alphas = np.logspace(-6, 6, 200)  # from 1e-6 to 1e6, 200 values in total

# Create RidgeCV object with specified alphas and number of folds (cv)
optimal_ridge2 = RidgeCV(alphas=ridge2_alphas, cv=10)

# Fit RidgeCV on training data to find the optimal alpha
optimal_ridge2.fit(X_2_train, y_2_train)

# Print the optimal alpha found
print('Optimal Alpha2:', optimal_ridge2.alpha_)
```

```python
# Initialize Ridge regression model with a specified alpha (regularization strength)
ridge2_model = Ridge(alpha=732)  

# Fit Ridge regression model
ridge2_model.fit(X_2_train, y_2_train)

# Predict on training and test sets
y_2_train_ridge_pred = ridge2_model.predict(X_2_train)
y_2_test_ridge_pred = ridge2_model.predict(X_2_test)

# Evaluate model for training and test sets
train_r2_ridge2 = r2_score(y_2_train, y_2_train_ridge_pred)
test_r2_ridge2 = r2_score(y_2_test, y_2_test_ridge_pred)

train2_rmse = root_mean_squared_error(y_2_train, y_2_train_ridge_pred)
test2_rmse = root_mean_squared_error(y_2_test, y_2_test_ridge_pred)

print('Train2 R2:', train_r2_ridge2)
print('Test2 R2:', test_r2_ridge2)

print('Train2 RMSE:', train2_rmse)
print('Test2 RMSE:', test2_rmse)

# Print the coefficients and intercept of the Ridge regression model
print("Ridge Coefficients:", ridge2_model.coef_)
print("Ridge Intercept:", ridge2_model.intercept_)
```
